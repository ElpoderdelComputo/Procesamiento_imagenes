```python
# üß† Procesamiento de im√°genes con OpenCV
# üé¨ Operaciones b√°sicas: redimensionar, rotar, recortar y convertir a escala de grises

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Leer imagen (BGR)
imagen = cv2.imread('/content/flor.png')

# Convertir a RGB para visualizar correctamente con matplotlib
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

plt.imshow(imagen_rgb)
plt.title('Imagen original')
plt.axis('off')
plt.show()
```

```python
# Mostrar tama√±o y propiedades de la imagen
print("Dimensiones de la imagen (alto, ancho, canales):", imagen.shape)
print("Tipo de dato de los p√≠xeles:", imagen.dtype)
```

```python
imagen.shape
```

```python
# Redimensionar a un tama√±o fijo
# redimensionada = cv2.resize(imagen_rgb, (tama√±o_x, tama√±o_y))
redimensionada = cv2.resize(imagen_rgb, (300, 100))

fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(imagen_rgb); axs[0].set_title('original (480, 817, 3)')
axs[1].imshow(redimensionada); axs[1].set_title('Redimensionada')
```

```python
# Redimensionar por factor
#peq_50 = cv2.resize(imagen_rgb, (x, y), factorX, factorY)
peq_50 = cv2.resize(imagen_rgb, (0, 0), fx=0.5, fy=0.5)

fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(imagen_rgb); axs[0].set_title('original')
axs[1].imshow(peq_50); axs[1].set_title('peque√±o')
```

```python
# Redimensionar a un tama√±o fijo
redimensionada = cv2.resize(imagen_rgb, (400, 300))

# Redimensionar por factor
peq = cv2.resize(imagen_rgb, (0, 0), fx=0.5, fy=0.5)
grande = cv2.resize(imagen_rgb, (0, 0), fx=1.5, fy=1.5)

fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(redimensionada); axs[0].set_title('400x300')
axs[1].imshow(peq); axs[1].set_title('Mitad del tama√±o')
axs[2].imshow(grande); axs[2].set_title('1.5x m√°s grande')
for ax in axs: ax.axis('off')
plt.show()
```

```python
# Rotaci√≥n
#print(imagen.shape)
alto, ancho = imagen.shape[:2]
centro = (ancho // 2, alto // 2)

# Matriz de rotaci√≥n (45 grados)
#matriz = cv2.getRotationMatrix2D(centro, grados_a_rotar, nivel_zoom)
matriz = cv2.getRotationMatrix2D(centro, 45, 1.0)
rotada = cv2.warpAffine(imagen_rgb, matriz, (ancho, alto))

fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(imagen_rgb); axs[0].set_title('Imagen original')
axs[1].imshow(rotada); axs[1].set_title('Imagen rotada')
plt.show()
```

```python
# Recorte
#recorte = imagen_rgb[y1:y2, x1:x2]
recorte = imagen_rgb[200:500, 300:700]
#recorte = imagen_rgb[150:500, 150:450] # mantener solo la primera flor

fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(imagen_rgb); axs[0].set_title('Imagen original')
axs[1].imshow(recorte); axs[1].set_title('Imagen recortada')
plt.show()
```

```python
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(imagen_rgb); axs[0].set_title('Imagen original')
axs[1].imshow(gris, cmap='gray'); axs[1].set_title('Escala de grises')
plt.axis('off')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0,0].imshow(imagen_rgb); axs[0,0].set_title('Original')
axs[0,1].imshow(redimensionada); axs[0,1].set_title('Redimensionada')
axs[1,0].imshow(rotada); axs[1,0].set_title('Rotada')
axs[1,1].imshow(gris, cmap='gray'); axs[1,1].set_title('Grises')

for ax in axs.flat:
    ax.axis('off')
plt.tight_layout()
plt.show()
```
